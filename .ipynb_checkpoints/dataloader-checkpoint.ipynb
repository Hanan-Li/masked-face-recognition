```python
import numpy as np
import pandas as pd
import os
import sys
from PIL import Image
import matplotlib.pyplot as plt

import pathlib
import tensorflow as tf
```

```python
path = "/Users/kevin/Desktop/Abroad/GaTech/master/Study/3rd_Semester/big/project/masked-face-recognition/"
data_path = path + "../data/"
face_path = data_path + "AFDB_face_dataset"
masked_face_path = data_path + "AFDB_masked_face_dataset"
```

```python
def preprocess_image(image):
  image = tf.image.decode_jpeg(image, channels=3)
  image = tf.image.resize(image, [150, 150])
  image /= 255.0  # normalize to [0,1] range

  return image
```

```python

def load_and_preprocess_image(path):
  image = tf.io.read_file(path)
  return preprocess_image(image)
```

```python
data_root = pathlib.Path(face_path)
data_root
```

```python
all_image_paths = list(data_root.glob('*/*'))
all_image_paths = [str(path) for path in all_image_paths]
image_count = len(all_image_paths)
all_image_paths[:5]
```

```python
label_names = sorted(item.name for item in data_root.glob('*/') if item.is_dir())
label_names[:5]
```

```python
label_to_index = dict((name, index) for index, name in enumerate(label_names))
```

```python
face_path
```

```python
all_image_labels = [label_to_index[pathlib.Path(path).parent.name]
                    for path in all_image_paths]

print("First 10 labels indices: ", all_image_labels[:10])
```

```python
ppath = "/Users/kevin/Desktop/Abroad/GaTech/master/Study/3rd_Semester/big/project/data/AFDB_face_dataset/aidai/0_0_aidai_0014.jpg"

load_and_preprocess_image(ppath)

img_path = ppath
label = "aidai"

plt.imshow(load_and_preprocess_image(img_path))
plt.grid(False)

plt.title(label)
print()
```

```python
path_ds = tf.data.Dataset.from_tensor_slices(all_image_paths)
AUTOTUNE = tf.data.experimental.AUTOTUNE
image_ds = path_ds.map(load_and_preprocess_image, num_parallel_calls=AUTOTUNE)
image_ds
```

```python
label_ds = tf.data.Dataset.from_tensor_slices(tf.cast(all_image_labels, tf.int64))
label_ds
```

```python
image_label_ds = tf.data.Dataset.zip((image_ds, label_ds))
image_label_ds
```

```python
mobile_net = tf.keras.applications.MobileNetV2(input_shape=(150, 150, 3), include_top=False)
mobile_net.trainable=False
```

```python
BATCH_SIZE = 32

# 设置一个和数据集大小一致的 shuffle buffer size（随机缓冲区大小）以保证数据
# 被充分打乱。
ds = image_label_ds.shuffle(buffer_size=image_count)
ds = ds.repeat()
ds = ds.batch(BATCH_SIZE)
# 当模型在训练的时候，`prefetch` 使数据集在后台取得 batch。
ds = ds.prefetch(buffer_size=AUTOTUNE)
ds
```

```python


ds = image_label_ds.apply(
  tf.data.experimental.shuffle_and_repeat(buffer_size=image_count))
ds = ds.batch(BATCH_SIZE)
ds = ds.prefetch(buffer_size=AUTOTUNE)
ds
```

```python
def change_range(image,label):
  return 2*image-1, label

keras_ds = ds.map(change_range)
```

```python
# 数据集可能需要几秒来启动，因为要填满其随机缓冲区。
image_batch, label_batch = next(iter(keras_ds))
```

```python
feature_map_batch = mobile_net(image_batch)
print(feature_map_batch.shape)
```

```python
model = tf.keras.Sequential([
  mobile_net,
  tf.keras.layers.GlobalAveragePooling2D(),
  tf.keras.layers.Dense(len(label_names), activation = 'softmax')])
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(),
              loss='sparse_categorical_crossentropy',
              metrics=["accuracy"])
```

```python
model.summary()
```

```python
steps_per_epoch=tf.math.ceil(len(all_image_paths)/BATCH_SIZE).numpy()
steps_per_epoch
```

```python
model.fit(ds, epochs=1, steps_per_epoch=steps_per_epoch)
```
